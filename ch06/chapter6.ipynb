### 6.1.2 SGD

* 重みパラメータ: $W$
* 勾配: $\frac{\partial L}{\partial W}$
* 学習係数: $\eta$

$$
  W \leftarrow W - \eta \frac{\partial L}{\partial W}
$$

### 6.1.4 Momentum

$$
  v \leftarrow \alpha v - \eta \frac{\partial L}{\partial W} \\
  W \leftarrow W + v
$$

### 6.1.5 AdaGrad

$$
  h \leftarrow h + \frac{\partial L}{\partial W} \odot \frac{\partial L}{\partial W} \\
  W \leftarrow W - \eta \frac{1}{\sqrt h} \frac{\partial L}{\partial W}
$$

$\odot$は行列の要素ごとの掛け算

## 6.3 Batch Normalization

### 6.3.1 Batch Normalizationのアルゴリズム

* 学習を早く進行させることができる（学習係数を大きくすることができる）
* 初期値にそれほど依存しない（初期値に対してそこまで神経質にならなくてよい）
* 過学習を抑制する（Dropoutなどの必要性を減らす）

具体的には、ミニバッチごとにデータの分布が平均0、分散1になるように正規化する。

## 6.4 正則化

### 6.4.2 Weight decay

* Weight decay（荷重減衰）: 重みが大きくなることを抑える
  * 過学習は重みパラメータが大きな値を取ることによって発生することが多い。

重みを$W$とすると$\frac{1}{2}\lambda W^2$を損失関数に加算する。（L2ノルム）

* L2ノルム: 各要素の2乗和
* L1ノルム: 各要素の絶対値の和
* L$\infty$ノルム: 各要素の絶対値の中で最大のもの
